```python
import cv2 as cv
```

```python
image = cv.imread('disease.jpg')
```

```python
leaf_cascade = cv.CascadeClassifier('cascade.xml')
image = cv.imread('disease.jpg')
g = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
leaf = leaf_cascade.detectMultiScale(g)
print(leaf)
```

```python
from PIL import Image, ImageDraw
im = Image.fromarray(g, mode='L')
dr = ImageDraw.Draw(im)
part = leaf.tolist()[0]
print(part)
dr.rectangle((part[0], part[1], part[0] + part[2], part[1] + part[3]), outline = 'red')
display(im)
```

```python
print(im.size)
```

```python
def shf(leaf):
    pil = Image.open('disease.jpg').convert('RGB')
    d = ImageDraw.Draw(pil)
    for x,y,w,h in leaf:
        d.rectangle((x,y,x+w,y+h), outline = 'orange')
    display(pil)
```

```python
leaf = leaf_cascade.detectMultiScale(cv.imread('disease.jpg'), 2.10)
print(leaf)
shf(leaf)
```

```python
def shf_h(leaf):
    pil = Image.open('healthy.jpg').convert('RGB')
    d = ImageDraw.Draw(pil)
    for x,y,w,h in leaf:
        d.rectangle((x,y,x+w,y+h), outline = 'orange')
    display(pil)
```

```python
leaf = leaf_cascade.detectMultiScale(cv.imread('healthy.jpg'), 2.10)
print(leaf)
shf_h(leaf)
```
